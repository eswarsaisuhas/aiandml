```python
## Importing libraries
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
## Loading the dataset from the sklearn
from sklearn.datasets import load_boston # importing the boston data from sklearn
boston = load_boston()
```

```python
boston_df = pd.DataFrame( boston.data, columns = boston.feature_names)
boston_df.head()
```

```python
boston_df['MEDV'] = boston.target
boston_df.head()
```

```python
## To find the correlation we use corr()
boston_df.corr()
```

```python
## Visulalizing the correlation
sns.heatmap(boston_df.corr())
```

```python
plt.figure(figsize=(14,14))
sns.heatmap(boston_df.corr(), annot=True, cmap='YlGnBu')
```

```python
## Correlation>0.7
correlation = boston_df.corr()
cor = correlation[correlation>0.7]
cor
```

```python
plt.figure(figsize=(10,10))
sns.heatmap(cor, cmap='Greens',annot=True)
```

```python
cor = correlation[(correlation>0.7) & (correlation>0.7)]
cor
```

```python
# Split the data into training and testing; Build the model on the training data and test the model on testing model
```

```python
## split the data into i/p and o/p
X = boston_df.drop(['MEDV'],axis=1)
y = boston_df.MEDV
```

```python
## Divide the X and y into train and test
from sklearn.model_selection import train_test_split
X_train, X_test, y_train,y_test = train_test_split(X, y, test_size=0.25, random_state=0)
```

```python
# Built the model on training data --> X_train and y_train
from sklearn.linear_model import LinearRegression
lm = LinearRegression()
lm.fit(X_train,y_train)
```

```python
# checking the Intercept and coefficient values
print(lm.intercept_)
lm.coef_
```

```python
## Create a dataframe for coefficients
coefficients = pd.DataFrame([X_train.columns, lm.coef_]).T
coefficients
```

```python
# Checking the model prediction on training data
y_train_pred = lm.predict(X_train)
y_train_pred
```

```python
# we need to compare the actual values(y_train) and the predicted values(y_train_pred)
from sklearn.metrics import r2_score
print('R^2: ',r2_score(y_train,y_train_pred))
print('Adjacent R^2: ',1-(1-r2_score(y_train,y_train_pred))*(len(X_train)-1)/(len(X_train)-X_train.shape[1]-1))
```

```python
# (alternate way)
numerator = (1-r2_score(y_train_pred,y_train))*(len(X_train)-1) 
denominator = (len(X_train)-X_train.shape[1]-1)

print("Adjusted R^2",1- (numerator/denominator))
```

```python
# regplot(y_train --> actual values and y_train_pred--> predicted values)
sns.regplot(y_train, y_train_pred)
```

```python
# Visualize the difference between actual and predicted values
plt.scatter(y_train,y_train_pred)
plt.xlabel('Price')
plt.ylabel('Model Predicted Price')
```

```python
## plot for residuals(errors)
## y_train-y_train_pred == Errors
plt.scatter(y_train_pred,y_train-y_train_pred)
plt.xlabel('Price')
plt.ylabel('Residulas')
```

```python
## Test the model on the testing data
y_test_pred=lm.predict(X_test) # test data--> unseen data
```
