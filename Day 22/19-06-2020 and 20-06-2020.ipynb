````python
#Imputing missing values
df.isnull().mean().sort_values(ascending=False)
````

````python
df.Cabin.head()
````

````python
## cabin-->77% missing values,we are going to create a new category as NA
df.Cabin.fillna('NA',inplace=True)
````

````python
df.Cabin.isnull().sum()
````

````python
df.Age.head()
````

````python
df.Age.describe()
````

````python
#median --> sorting in the order (middle value) NaN in the age column should be substituted with median of age
df.Age.fillna(df.Age.median(),inplace=True)
df.Age.isnull().sum()
````

````python
df.Embarked.value_counts()
````

````python
df.Embarked.mode()
````

````python
#imputing embarked column with mode 
df.Embarked.fillna(df.Embarked.mode()[0],inplace=True)
df.Embarked.isnull().sum()
````

````python
#imputing farecategory column
df.Fare_Category.isnull().sum()
````

````python
df.Fare_Category.fillna(df.Fare_Category.mode()[0],inplace=True)
df.Fare_Category.isnull().sum()
````

````python
df.isnull().sum()
````

````python
sns.heatmap(df.isnull())
````

````python
df.head()
````

````python
df['Age-Range'] = pd.cut(df.Age,[0,10,20,30,40,60,80])
df[['Age','Age-Range']].head()
````

````python
dummy_set = pd.get_dummies(df.Fare_Category,drop_first=True)
dummy_set # n columns,we need to include upto n-1 dummy columns
````

````python
df = pd.concat([df,dummy_set],axis=1)
````

````python
# Encoding of categorical variables/columns
df = pd.concat([df,pd.get_dummies(df.Cabin,prefix='Cabin',drop_first=True),
               pd.get_dummies(df.Embarked,prefix='Emb',drop_first=True),
               pd.get_dummies(df.Fare_Category,prefix='Fare',drop_first=True),
               pd.get_dummies(df['Age-Range'],drop_first=True)],axis=1)
df.head()
````

````python
#encoding sex column
from sklearn.preprocessing import LabelEncoder
df['Sex'] = LabelEncoder().fit_transform(df.Sex)
df.head(1)
````

````python
df.drop(['PassengerId','Name','Age','SibSp','Parch','Ticket','Fare',
         'Embarked','Fare_Category','Age-Range','Cabin'],axis=1,inplace=True)
df.shape
````

````python
df.head()
````

````python
df.dtypes
````

````python
#StandardScalar --> not required for this(DistanceBased Algorithm)
````

````python
## separating the input and output data
X = df.drop('Survived',axis=1)
X.shape
````

````python
# output column
y = df.iloc[:,0]
y.shape
````

````python
# input and output into training and test data
#training --> we will betraining the model on training data
#testing --> check the performance of model
````

````python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.25,random_state=1)#random state to select constant rows
````

## For Classification: we are classifying whether the passenger is survived or not
- Logistic Regression

````python
from sklearn.linear_model import LogisticRegression
log_reg = LogisticRegression() # creating an object for Logistic Regression
````

````python
# we have to apply this object(Log_reg) to the training data
log_reg.fit(X_train,y_train) # with the help of fit method we are fitting logistic regression with training data

##objectName.fit(InputData, outputData)
````

### Instead of directly predicting on test data, let us see how well the model predicts the training data

- Predicting on the training data

```syntax: objectName.predict(Input)```

````python
y_train_pred = log_reg.predict(X_train)
````

- y_train --> original output
- y_train_pred --> model predicted o/p

````python
y_train == y_train_pred # comparing original data o/p and model predicted o/p
````

````python
## accuracy_score --> with the help of the metric, we can evaluate the overall performace of the model(train data)
from sklearn.metrics import accuracy_score
accuracy_score(y_train, y_train_pred)
````

## performance Metrics --> To evaluate the model is predecting

````python
# Predicting the model on text data
y_test_pred = log_reg.predict(X_test)
````

````python
y_train_pred
````

````python
y_test == y_test_pred
````

````python
# checking the accuracy_score of the test data 
accuracy_score(y_test, y_test_pred)
````
